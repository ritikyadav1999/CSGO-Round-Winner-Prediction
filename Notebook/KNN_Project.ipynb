```python
import pandas as pd
```

```python
csgo = pd.read_csv("https://raw.githubusercontent.com/insignificantGuy/Kaggle-Datasets/main/csgo_round_snapshots.csv")
```

```python
csgo.head(10)
```

```python
csgo.info()
```

```python
csgo.isna().sum()
```

```python
df = pd.get_dummies(data=csgo,drop_first=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
array = scaler.fit_transform(df.drop('round_winner_T',axis=1))
dif = pd.DataFrame(array,columns=df.columns[:-1])
```

```python
dif
```

```python
x = dif
y = df['round_winner_T']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state = 100)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knc = KNeighborsClassifier()
knc.fit(x_train,y_train)
```

```python
predict = knc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
matrix = confusion_matrix(y_test,predict)
matrix
```

```python
print(classification_report(y_test,predict))
```
